# Splitting the dataset

```python
dataset.head()
```

```python
# independent / Matrix of features
x = dataset.drop(columns=['entry_id', 'e_signed'])
```

```python
# target/dependent variable
y = dataset['e_signed']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=0 )
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Part 2: Building the Model

# 1) Logistic regression

```python
from sklearn.linear_model import LogisticRegression

classifier_lr = LogisticRegression(random_state=0)
classifier_lr.fit(x_train, y_train)
```

```python
y_pred = classifier_lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

# 2) SVM

```python
from sklearn.svm import SVC

classifier_svc = SVC(random_state=0)
classifier_svc.fit(x_train, y_train)
```

```python
y_pred = classifier_svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

# Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier

classifier_rf = RandomForestClassifier(random_state=0)
classifier_rf.fit(x_train, y_train)
```

```python
y_pred = classifier_rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

# XGBoost Classifier

```python
from xgboost import XGBClassifier

classifier_xgb = XGBClassifier(random_state=0)
classifier_xgb.fit(x_train, y_train)
```

```python
y_pred = classifier_xgb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

# Part 3: Applying Randomized Search to find the best parameters

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {
    'learning_rate': [0.05, 0.10, 0.15, 0.20, 0.25, 0.30],
    'max_depth': [2, 3, 5, 10, 15],
    'min_child_weight': [1, 3, 5, 7],
    'gamma': [0.0, 0.1, 0.2, 0.3, 0.4],
    'colsample_bytree': [0.3, 0.4, 0.5, 0.6, 0.7],
    'n_estimators': [100, 200, 300, 400, 500],
    'subsample': [0.5, 0.7, 1.0]
}
```

```python
parameters
```

```python
random_cv = RandomizedSearchCV(
    estimator=classifier_xgb,
    param_distributions=parameters,
    n_iter=5,
    scoring='roc_auc',
    n_jobs=-1,
    cv=5,
    verbose=3
)
```

```python
random_cv.fit(x_train, y_train)
```

```python
random_cv.best_estimator_
```

```python
random_cv.best_params_
```

```python
random_cv.best_score_
```

# Part 4: Final model (XGBoost Classifier)

```python
from xgboost import XGBClassifier

classifier = XGBClassifier(
    colsample_bytree=0.3,
    learning_rate=0.15,
    max_depth=5,
    n_estimators=100,
    gamma=0.0,
    min_child_weight=1,
    missing=np.nan )

classifier.fit(x_train, y_train)
```
